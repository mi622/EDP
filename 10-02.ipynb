Ejercicio 1. Tomar a m= $2^7$ y elegir los parametros adecuados para obtener una muestra de numeros aleatorios con ciclo de vida completo.

a) Guardar los numeros en una lista y obtener su media

b)Graficar el histograma de la muestra

```python

import tabulate
```

```python

a=5
c=1
m=128
x0=1
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generador(a,c,m,x0):
  x1=(a*x0+c)  %m
  return x1
```

```python
generador(a,c,m,x0)
```

```python
muestra=[]
datos=[]
for i in range(130):
  nuevo = generador(a,c,m,x0)
  muestra.append([i,x0,nuevo/m])
  datos.append(x0/m)
  x0=nuevo
```

```python

muestra
```

```python
datos
```

```python
np.mean(datos) #calcula la media
```

```python
np.var(datos)
```
