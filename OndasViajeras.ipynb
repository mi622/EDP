Gráfique la función
$$ f(x')=exp ( - \frac{(x'-0.5)^2}{0.01}) $$
y suponga que esta representa una turbación en una cuerda.

```python
# prompt: Gráfique la función
# $$ f(x')=exp ( - \frac{(x'-0.5)^2}{0.01}) $$
# y suponga que esta representa una turbación en una cuerda.

import numpy as np
import matplotlib.pyplot as plt

def f(x_prime):
  return np.exp(-(x_prime - 0.5)**2 / 0.01)

x_prime = np.linspace(0, 1, 100)
y = f(x_prime)

plt.plot(x_prime, y)
plt.title("Turbación en una cuerda")
plt.xlabel("x'")
plt.ylabel("f(x')")
plt.grid(True)
plt.show()
```

(a) Mostrar que al hacer el cambio de variable x'= x-ct la perturbación se desplaza a la derecha.

```python

c = 0.1  # Velocidad de propagación (arbitraria)
times = [0,1,2,3] # Diferentes instantes de tiempo

plt.figure(figsize=(10, 6))

for t in times:
  x = np.linspace(0, 2, 100) # Definimos un rango más amplio para 'x'
  x_prime_at_t = x - c * t
  y_at_t = f(x_prime_at_t)
  plt.plot(x, y_at_t, label=f't = {t}')

plt.title("Desplazamiento de la turbación con el tiempo")
plt.xlabel("x")
plt.ylabel("f(x - ct)")
plt.legend()
plt.grid(True)
plt.show()

```

(b) Mostrar que al hacer el cambio de variable x'= x+ct la perturbación se desplaza a la izquierda.

```python
c = 0.1  # Velocidad de propagación (arbitraria)
times = [0,1,2,3] # Diferentes instantes de tiempo

plt.figure(figsize=(10, 6))

for t in times:
  x = np.linspace(0, 2, 100) # Definimos un rango más amplio para 'x'
  x_prime_at_t = x + c * t
  y_at_t = f(x_prime_at_t)
  plt.plot(x, y_at_t, label=f't = {t}')

plt.title("Desplazamiento de la turbación con el tiempo")
plt.xlabel("x")
plt.ylabel("f(x - ct)")
plt.legend()
plt.grid(True)
plt.show()
```

(c) Graficar la superposición de las ondas viajeras de los incisos (a) y (b).

```python

c = 0.1  # Velocidad de propagación (arbitraria)
times = [0, 1, 2, 3] # Diferentes instantes de tiempo

plt.figure(figsize=(10, 6))

for t in times:
  x = np.linspace(0, 2, 100) # Definimos un rango más amplio para 'x'
  x_prime_at_t_right = x - c * t
  y_at_t_right = f(x_prime_at_t_right)

  x_prime_at_t_left = x + c * t
  y_at_t_left = f(x_prime_at_t_left)

  y_superposition = y_at_t_right + y_at_t_left

  plt.plot(x, y_superposition, label=f't = {t}')

plt.title("Superposición de Ondas Viajeras")
plt.xlabel("x")
plt.ylabel("f(x - ct) + f(x + ct)")
plt.legend()
plt.grid(True)
plt.show()
```
